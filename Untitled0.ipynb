```python
import pandas as pd
df = pd.read_excel("student_feedback.xlsx")
df.head()
```

```python
# Drop duplicate rows
df = df.drop_duplicates()

# Fill missing numeric values with the median of each column
df = df.fillna(df.median(numeric_only=True))

# Make sure all rating columns (except Student ID) are numeric
rating_cols = df.columns[1:]  # exclude Student ID
df[rating_cols] = df[rating_cols].apply(pd.to_numeric, errors='coerce')

# Confirm cleaning
print("Missing values after cleaning:\n", df.isnull().sum())
df.head()
```

```python
# Keep Student ID separately
student_col = "Student ID"

# All feedback columns (numeric ratings)
feedback_cols = df.columns.drop(student_col)

# Ensure ratings are numeric
df[feedback_cols] = df[feedback_cols].apply(pd.to_numeric, errors='coerce')

# Clip to valid range 1–10 (in case of wrong entries)
df[feedback_cols] = df[feedback_cols].clip(lower=1, upper=10)

# Prepared dataset: Student IDs + Cleaned Ratings
prepared_df = df[[student_col] + list(feedback_cols)]

# Quick check
prepared_df.head()
```

```python
summary = prepared_df[feedback_cols].describe().T
summary[["mean", "std", "min", "max"]]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

avg_ratings = prepared_df[feedback_cols].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_ratings.values, y=avg_ratings.index, palette="viridis")
plt.title("Average Ratings by Feedback Criteria", fontsize=14, weight="bold")
plt.xlabel("Average Score (1–10)")
plt.ylabel("Feedback Criteria")
plt.show()
```

```python
def rating_to_sentiment(score):
    if score <= 4:
        return "Negative"
    elif score <= 7:
        return "Neutral"
    else:
        return "Positive"

# Apply to all feedback columns
sentiment_df = prepared_df.copy()
for col in feedback_cols:
    sentiment_df[col] = prepared_df[col].apply(rating_to_sentiment)

sentiment_df.head()
```

```python
plt.figure(figsize=(12,8))
for i, col in enumerate(feedback_cols, 1):
    plt.subplot(3, 3, i)
    sns.histplot(prepared_df[col], bins=10, kde=True, color="blue")
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(prepared_df[feedback_cols].corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Between Feedback Criteria", fontsize=14, weight="bold")
plt.show()
```

```python
avg_ratings = prepared_df[feedback_cols].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
ax = sns.barplot(x=avg_ratings.values, y=avg_ratings.index, palette="coolwarm")

# Add criteria labels inside bars
for i, (value, name) in enumerate(zip(avg_ratings.values, avg_ratings.index)):
    ax.text(value/2, i, name, ha='center', va='center', color='white', fontsize=10, weight='bold')

# Remove y-axis labels (since inside bars now)
ax.set_yticklabels([])

# Threshold lines
plt.axvline(7, color="red", linestyle="--", label="Needs Improvement Threshold")
plt.axvline(8, color="green", linestyle="--", label="Strong Performance")

# Legend outside
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title("Strengths and Improvement Areas", fontsize=14, weight="bold")
plt.xlabel("Average Score (1–10)")
plt.ylabel("")  # remove y-axis title
plt.tight_layout()
plt.show()
```
